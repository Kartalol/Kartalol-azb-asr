```python
from transformers import AutoModelForCTC, Wav2Vec2BertProcessor
import torch
import librosa

model = AutoModelForCTC.from_pretrained("./model").to("cuda")
processor = Wav2Vec2BertProcessor.from_pretrained("./model")
```

```python
def infer(file: str):
    audio, sr = librosa.load(
        file,
        sr=16000,
    )
    input_values = processor(audio, sampling_rate=sr, return_tensors="pt").to(
        model.device
    )
    # ins = torch.tensor(input_values["input_features"]).to("cuda").unsqueeze(0)
    logits = model(**input_values).logits
    pred_ids = torch.argmax(logits, dim=-1)[0]

    return processor.decode(pred_ids)


infer("/home/davud/Desktop/storage/webspeech/azeri-test/audio/f2_ID35_492_book1.wav")
```
